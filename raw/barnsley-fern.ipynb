```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_barnsley(A):

    df = pd.DataFrame([[np.array([[row[0], row[1]], [row[2], row[3]]]), 
                        np.array([[row[4]], [row[5]]]), 
                        row[6]] for row in A],
                     columns=["A", "B", "p"])

    n = 100000
    pts = np.empty((n, 2))
    pts[0] = [0, 0]

    choices = np.random.choice(len(A), n, p=df['p'])

    for row in range(1, n):
        i = choices[row]
        pts[row]=(np.matmul(df["A"][i], pts[row-1]).reshape((2, 1))+df["B"][i]).reshape((1, 2))

    plt.scatter(*zip(*pts), s=0.1, c='g')
```

Here is the classic Barnsley fern resembling the black spleenwort, _Asplenium adiantum-nigrum_. 

```python
A = np.array([[0, 0, 0, 0.16, 0, 0, 0.01],
    [0.85, 0.04, -0.04, 0.85, 0, 1.6, 0.85],
    [0.2, -0.26, 0.23, 0.22, 0, 1.6, 0.07],
    [-0.15, 0.28, 0.26, 0.24, 0, 0.44, 0.07]])

generate_barnsley(A)
```

This one's supposed to look like a _Cyclosorus_ or _Thelypteridaceae_ fern.

```python
B = np.array([
    [0, 0, 0, 0.25, 0, -0.4, 0.02],
    [0.95, 0.005, -0.005, 0.93, -0.002, 0.5, 0.84],
    [0.035, -0.2, 0.16, 0.04, -0.09, 0.02, 0.07],
    [-0.04, 0.2, 0.16, 0.04, 0.083, 0.12, 0.07]
])

generate_barnsley(B)
```
